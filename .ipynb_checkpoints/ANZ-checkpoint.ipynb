```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
pd.options.display.max_columns=None
pd.options.display.max_rows=None
```

```python
df=pd.read_excel('ANZ synthesised transaction dataset.xlsx')
df.head()
```

```python
df['longitude']=df['long_lat'].apply(lambda x : float(x.split('-')[0]))
df['latitude']=df['long_lat'].apply(lambda x : float(x.split('-')[1]))
df.head()
```

```python
df['merchant_longitude']=df['merchant_long_lat'].apply(lambda x : x.split('-')[0])
df['merchant_latitude']=df['merchant_long_lat'].apply(lambda x : x.split('-')[1])
df.head()
```

```python
df.info()
```

```python
df.isnull().mean()
```

```python
irr_col=[var for var in df.columns if var in df.columns[df.isnull().mean()>0.9]]
print(irr_col)
df.drop(irr_col, axis=1, inplace=True)
```

Categorical Columns

```python
cat_var=[var for var in df.columns if var in df.columns[df.dtypes=='object']]
cat_var
```

Numeric columns 

```python
num_var=[var for var in df.columns if var in df.columns[df.dtypes!='object']]
num_var
```

```python
def cardinality(df, col_list):
    for var in col_list:
        print(var,df[var].value_counts()/df.shape[0]*100,'\n')
    
cardinality(df, cat_var)
```
